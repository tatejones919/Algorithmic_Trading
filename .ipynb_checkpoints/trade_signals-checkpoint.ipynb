```python
from alpha_vantage.timeseries import TimeSeries
import pandas as pd
import numpy as np
from pathlib import Path
import hvplot.pandas
import quandl
import requests
```

```python
url_eur= 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=EURUSD&outputsize=full&apikey='+'8YH9QLMWXNG5DNG0'
r_eur = requests.get(url_eur)
eur_usd = r_eur.json()
#eur_usd['Time Series (Daily)']
```

```python
#app = TimeSeries('8YH9QLMWXNG5DNG0')
#eur_usd = app.get_daily_adjusted('EURUSD', outputsize=full)
```

```python
'''
date= eur_usd[0].keys()
close = []
for item in date:
    close.append(eur_usd[0][item]['4. close'])
'''
```

```python
date= eur_usd['Time Series (Daily)'].keys()
close = []
for item in date:
    close.append(eur_usd['Time Series (Daily)'][item]['4. close'])
```

```python
eur_usd = pd.DataFrame(close, columns = ['Close'])
eur_usd['Date']= date
eur_usd.set_index(['Date'], inplace=True)
eur_usd.sort_values(by='Date', ascending = True, inplace=True)
eur_usd = eur_usd['2018-09-21':]
eur_usd.reset_index(inplace= True)
eur_usd
```

```python
signals_df = eur_usd.loc[:, ["Date", "Close"]].copy()

# Set the short window and long windows
short_window = 50
long_window = 100

# Set the `date` column as the index
signals_df = signals_df.set_index("Date", drop=True)

# Generate the short and long moving averages (50 and 100 days, respectively)
signals_df["SMA50"] = signals_df["Close"].rolling(window=short_window).mean()
signals_df["SMA100"] = signals_df["Close"].rolling(window=long_window).mean()
signals_df["Signal"] = 0.0

# Generate the trading signal 0 or 1,
# where 0 is when the SMA50 is under the SMA100, and
# where 1 is when the SMA50 is higher (or crosses over) the SMA100
signals_df["Signal"][short_window:] = np.where(
    signals_df["SMA50"][short_window:] > signals_df["SMA100"][short_window:], 1.0, 0.0
)

# Calculate the points in time at which a position should be taken, 1 or -1
signals_df["Entry/Exit"] = signals_df["Signal"].diff()

```

```python
signals_df.head()
```

```python
signals_df["Close"] = pd.to_numeric(signals_df["Close"], downcast="float")
```

```python

exit = signals_df[signals_df['Entry/Exit'] == -1.0]['Close'].hvplot.scatter(
    color='red',
    marker='v',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

entry = signals_df[signals_df['Entry/Exit'] == 1.0]['Close'].hvplot.scatter(
    color='green',
    marker='^',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

security_close = signals_df[['Close']].hvplot(
    line_color='lightgray',
    ylabel='Price in $',
    width=1000,
    height=400
)

moving_avgs = signals_df[['SMA50', 'SMA100']].hvplot(
    ylabel='Price in $',
    width=1000,
    height=400
)
entry_exit_plot = security_close * moving_avgs * entry * exit
entry_exit_plot.opts(xaxis=None)
```

# Creating Various Trade Signlas

```python
eur_usd.set_index(pd.to_datetime(eur_usd['Date'], infer_datetime_format=True), inplace=True)
eur_usd.drop(columns=['Date'], inplace=True)
eur_usd['Close'] = eur_usd['Close'].astype(float)
eur_usd
```

```python
eur_usd['Daily Returns'] = eur_usd['Close'].dropna().pct_change()
eur_usd
```

```python
# Set short and long windows
short_window_eur = 1
long_window_eur = 10

# Construct a `Fast` and `Slow` Exponential Moving Average from short and long windows, respectively
eur_usd['fast_close'] = eur_usd['Close'].ewm(halflife=short_window).mean()
eur_usd['slow_close'] = eur_usd['Close'].ewm(halflife=long_window).mean()

# Construct a crossover trading signal
eur_usd['crossover_long'] = np.where(eur_usd['fast_close'] > eur_usd['slow_close'], 1.0, 0.0)
eur_usd['crossover_short'] = np.where(eur_usd['fast_close'] < eur_usd['slow_close'], -1.0, 0.0)
eur_usd['crossover_signal'] = eur_usd['crossover_long'] + eur_usd['crossover_short']

eur_usd.head()
```

```python
eur_usd[['Close', 'fast_close', 'slow_close']].plot(figsize=(20,10))
```

```python
# Set short and long volatility windows
short_vol_window_eur = 1
long_vol_window_eur = 10

# Construct a `Fast` and `Slow` Exponential Moving Average from short and long windows, respectively
eur_usd['fast_vol'] = eur_usd['Daily Returns'].ewm(halflife=short_vol_window_eur).std()
eur_usd['slow_vol'] = eur_usd['Daily Returns'].ewm(halflife=long_vol_window_eur).std()

# Construct a crossover trading signal
eur_usd['vol_trend_long'] = np.where(eur_usd['fast_vol'] < eur_usd['slow_vol'], 1.0, 0.0)
eur_usd['vol_trend_short'] = np.where(eur_usd['fast_vol'] > eur_usd['slow_vol'], -1.0, 0.0) 
eur_usd['vol_trend_signal'] = eur_usd['vol_trend_long'] + eur_usd['vol_trend_short']

eur_usd
```

```python
eur_usd[['fast_vol', 'slow_vol']].plot(figsize=(20,10))
```

```python
# Set bollinger band window
bollinger_window_eur = 20

# Calculate rolling mean and standard deviation
eur_usd['bollinger_mid_band'] = eur_usd['Close'].rolling(window=bollinger_window_eur).mean()
eur_usd['bollinger_std'] = eur_usd['Close'].rolling(window=20).std()

# Calculate upper and lowers bands of bollinger band
eur_usd['bollinger_upper_band']  = eur_usd['bollinger_mid_band'] + (eur_usd['bollinger_std'] * 1)
eur_usd['bollinger_lower_band']  = eur_usd['bollinger_mid_band'] - (eur_usd['bollinger_std'] * 1)

# Calculate bollinger band trading signal
eur_usd['bollinger_long'] = np.where(eur_usd['Close'] < eur_usd['bollinger_lower_band'], 1.0, 0.0)
eur_usd['bollinger_short'] = np.where(eur_usd['Close'] > eur_usd['bollinger_upper_band'], -1.0, 0.0)
eur_usd['bollinger_signal'] = eur_usd['bollinger_long'] + eur_usd['bollinger_short']

eur_usd
```

```python
# Plot the Bollinger Bands for BTC/USD closing prices
eur_usd[['Close','bollinger_mid_band','bollinger_upper_band','bollinger_lower_band']].plot(figsize=(20,10))
```
